```python
data_subset_1 = data[['like','share']] #creating data subset/ data frame
data_subset_1
```

```python
data_subset_2 = data[['comment','Type']] #creating another data subset
data_subset_2
```

```python
merged_data  = pd.merge(data_subset_2, data_subset_1, left_on='comment', right_on= 'like') #merging the two subsets
merged_data
```

```python
merged_data.sort_values(by=['like'],ascending=False) #sorting the merged data
```

```python
merged_data.transpose() #transposing the merged data
```
